```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.describe()
df
```

```python
graph = sns.boxplot(x='Survived', y='Fare', data=df)
graph.set(ylim=(0, 125))
```

```python
graph = sns.lineplot(x='Age', y='Fare', hue='Survived', data=df)
```

```python
graph = sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
graph.set(ylim=(0, 125))
```
